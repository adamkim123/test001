## `pymatgen.core.structure`

### `structure. `

```python
from pymatgen.core import Structure
```

```python
# Reading a cif file 
cif_file = "high_density_defects/MoS2_500/highMoS2cifs/MoS2_Mo56W4Se2S122_61d5e59e-335f-4026-bf18-15769c43a20b.cif"
structure = Structure.from_file(cif_file)

# print(structure)
```

```python
for i in structure:
    print(i)
```

```python
# Get number of sites
print(structure.num_sites )
```

```python
# Get types of species
print(structure.types_of_specie)
```

```python
# Get number of species
print(len(structure.types_of_specie))
```

```python
# Get atomic number of the type of specie
print(f"{i}:{i.Z for i in structure.types_of_specie})
```

```python
# Get the chemical formula indicating total sites per element 
print(structure.formula)
```

```python
# Get reduced chemical formula
print(structure.reduced_formula)
```

```python
# Get all the species in the structure
print(structure.species)
```

```python
# Extract the cartesian coordinates of every site
print(structure.cart_coords)
```

```python
# Get fractional coordinates of every site
print(structure.frac_coords)
```

```python
# Get the atomic numbers of the sepcies in the structure
print(structure.atomic_numbers) # This produces a tuple

# To change from tuple to  list, you can use list()

sample = list(structure.atomic_numbers)
print(sample)
```

```python
# Get the combined element type, fractional and cartecian coordinates
print(structure.sites)
```

```python
# What can you do with a site
for i in structure.sites[0]:
    print(i.specie) # the element involved
    print(i.specie.Z) # The atomic number of element involved
    print(i.frac_coords) # Can get the fractional coords
```

```python
# Get the occupancy value of each specie
print(structure.species_and_occu)
```

```python
for x in structure:
    print(f"{x.specie}:{x.frac_coords}")
```

```python
sam = structure.make_supercell
print(sam)
```

### `pymatgen.core.Structure.lattice`
This contains methods used to find the general information of the structure.

```python
# Extract lattice vectors as list
lat = structure.lattice
print(lat)

# Turn this into a matrix
lat_matrix = structure.lattice.matrix
print(lat_matrix)
```

```python
# Extract lattice abc
print(f"abc values: {structure.lattice.abc}")
print(f"a: {structure.lattice.a}")
print(f"b: {structure.lattice.b}")
print(f"c: {structure.lattice.c}")
```

```python
# Get volume of the unit cell
print(structure.lattice.volume)
```

```python
print(structure.lattice.pbc)
```

```python
# Get angles
print(f"Angles: {structure.lattice.angles}")
print(f"Alpha: {structure.lattice.alpha}")
print(f"Beta: {structure.lattice.beta}")
print(f"Gamma: {structure.lattice.gamma}")
```

```python
print(structure.lattice.get_cartesian_coords)
```

```python
# Get space group information
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
spg = SpacegroupAnalyzer(structure)
space_group = spg.get_space_group_symbol()
print(space_group)
```

### `pymatgen.core.Lattice`

```python
from pymatgen.core import Lattice
```

```python
# Silicon cubic lattice
lattice = Lattice.cubic(5.431)  # in cubic lattice, a = b = c
species = ["Si", "Si"]
coords = [[0, 0, 0], [0.25, 0.25, 0.25]]  # Diamond structure # Find coordinates of a unit cell

# Create Silicon unit cell
si_unit_cell = Structure(lattice, species, coords)

print(si_unit_cell)
```

```python
# Define supercell scaling (2x2x2 expansion)
# these are the vector coordinates
supercell = si_unit_cell.make_supercell([[2, 0, 0], [0, 2, 0], [0, 0, 2]])

print("\nSupercell:")
print(supercell)
```

## `pymatgen.analysis`

This is used to convert cif files into crystal graphs
A graph has **nodes** (which will be the atoms) and **edges** (which is the bonds between the atoms)

```python
from pymatgen.analysis.graphs import StructureGraph
from pymatgen.analysis.local_env import MinimumDistanceNN, CrystalNN # For strategy selection
```

```python

# Create a graph based on nearest neighbors
graph_1 = StructureGraph.with_local_env_strategy(structure, MinimumDistanceNN())
graph_1
```

```python
graph_2 = StructureGraph.with_local_env_strategy(structure, CrystalNN())
graph_2
```

```python
# Get coordination number(local environment)
cnn = CrystalNN()

coordination_numbers = [cnn.get_cn(structure,i) for i in range(len(structure))]
coordination_numbers
```

```python
# Get the nearest neighbors
nearest_neighbor = [cnn.get_nn(structure,i) for i in range (len(structure))]
nearest_neighbor
```
